```python
#11[4]
array = np.random.randint(-100, 100, size=20)
def merge(array, left, right):
  a = np.partition(left, len(left)-1)
  b = np.partition(right, len(right)-1)
  arr = np.hstack([a, b])
  arr = np.partition(arr, np.arange(len(array)))
  return arr

def mergeSort(array):
  left, right = np.split(array, [2])
  if len(left) > 1 and len(right) > 1:
    mergeSort(left), mergeSort(right) 
  return merge(array, left, right)

print(mergeSort(array))
```
