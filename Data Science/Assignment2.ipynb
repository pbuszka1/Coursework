```python
import numpy as np
import pandas as pd
```

```python
#Q1
data = pd.Series()
x = 2
count = 0
def is_prime(number):
  for i in range(2, number):
    if number % i == 0:
      return False
  else: 
      return True

while count < 10:
  if is_prime(x) == True:
    temp = pd.Series([x], index=[count])
    temp2 = pd.concat([data, temp])
    data = temp2
    x+=1
    count+=1
  else:
    x+=1

data
```
