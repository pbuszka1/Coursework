```python
#Q4
student = pd.DataFrame({'student_name':['John', 'Bob', 'Suzan'], 'age':[20,30,22]})
course = pd.DataFrame({'department name':['CS','CS', 'ENGL'],'course number':[233,455, 433], 'description':['art of Stackoverflow', 'how to google', 'pointless']})
takes = pd.DataFrame({'department name':['CS','CS', 'ENGL'],'course number':[233,455, 433], 'grade':[[2, 3, 0], [0, 0, 0], [0, 0, 4]]})
print(student)
print(course)
print(takes)
```

```python
#Q5
df = pd.merge(student, takes, left_index=True, right_index=True)
df = df.set_index(['student_name'])
def mean(x):
  count = 0
  for i in range(len(x)):
    if x[i] != 0:
      count += 1
  if count == 0:
    return 0
  else:
    return sum(x) / count
df = df['grade'].apply(lambda x: mean(x[:3]))
df
```

```python
#Q8
def days_count(day):
  return len(pd.date_range(start='2018-01-01', end='2018-12-31', freq='W-'+ day))

days = ['MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT', 'SUN']
num = pd.DataFrame([days_count(i) for i in days], index=days)
num
```

```python
#Q9
google_data = yf.download('GOOG', start='2017-01-01', end='2017-12-31')
google_data['Profit']=google_data['Close']-google_data['Open']
google_data = google_data.reset_index()
google_data['DOW'] = google_data['Date'].transform(lambda x: x.strftime('%A'))
result = google_data.groupby('DOW')['Profit'].sum() * 20 # there was stock split
pd.DataFrame(result, columns=['Profit']).sort_values('Profit', ascending=False)
```
