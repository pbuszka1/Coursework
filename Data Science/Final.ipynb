```python
#3
Movie = pd.DataFrame({'name':['Star Wars', 'Harry Potter', 'Any Adam Sandler Movie'], 'year_made':[1977, 2001, 1980]})
Room = pd.DataFrame({'room_num':[1,2,3], 'size':[50,70,100]})
Projection = pd.DataFrame({'projection_id':[1000, 1002, 1004], 'date':['5/1/77', '8/13/01', '1/1/80'], 'time':['11:00','17:00','5:00'], 'name':['Star Wars', 'Harry Potter', 'Any Adam Sandler Movie']})
Ticket = pd.DataFrame({'projection_id':[1000, 1002, 1004], 'tickets':[7, 5, 1], 'price':[7,5,1]})
print(Movie, Room, Projection, Ticket)
```

```python
#4 
Ticket = Ticket.join(Room)
Ticket['occ_rate'] = Ticket['tickets'] / Room['size']
Ticket['occ_rate'].max()
Ticket['room_num']
```

```python
#5
Movie = Movie.join(Ticket)
Movie['most_money'] = Ticket['tickets'] * Ticket['price']
Movie['most_money'].max()
Movie['name'][0]
```

```python
#6
Movie['most_money'].max()
Movie = Movie.merge(Projection, on=['projection_id', 'name'])
Movie['date'][0]
```

```python
#7
Movie['date'].min()
Movie['name'][0]
```
